# 1. Analysing Number of matches won by a each team in t-20 world cup

```python
win_count=data.groupby('winner')['winner'].count().sort_values(ascending=False)
print(win_count)
```

```python
win_count.plot(kind='bar',figsize=(20,5),title='Number of Matches Won by teams in t20 World',
               xlabel='Teams',ylabel='No of matches won',rot=0)
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=data['winner'])
```

__As we can see England has won the world cup as they won the five matches and india and pakistan won the four mactches respectively.__

# 2.Analysing How many matches won by wickets and runs.

```python
won_by=data.groupby('won by')['won by'].count()
print(won_by)
```

```python
plt.pie(won_by,labels=['Runs','Wickets'],startangle=100,explode=[0,0.1],shadow=True,colors=['r','m'],radius=1)
plt.title('Number of Matches Won By Runs Or Wickets')
plt.legend(loc=1)
plt.show()
```

__So in the t-20 worldcup 2022, 16 matches were won by batting first and 13 matches were won by chasing.__

# 3.Analysing What was the toss decision by teams in t20 worldcup 2022.

```python
toss_decision=data['toss decision'].value_counts()
print(toss_decision)
```

```python
plt.pie(toss_decision,labels=['Bat first','field first'],startangle=100,explode=[0.1,0],shadow=True,colors=['Cyan','Yellow'])
plt.title('Toss Decisions in t20 World Cup 2022')
plt.legend(loc=1)
plt.show()
```

__So we can see in the t20 worldcup 2022 in 17 matches team chose to bat first and in 13 matches chose field first.__

# 4.Analysing Highest run getters in t20 worldcup 2022.

```python
top_scorers=data.groupby('top scorer')['highest score'].sum().sort_values(ascending=False)
print(top_scorers)
```

```python
top_scorers.plot(kind='bar',figsize=(20,5),rot=40,title='Top Scorers in t20 World Cup 2022',\
                 xlabel='Top Scorer',ylabel='Higest Score',color='g')
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation=45)
sns.barplot(x=top_scorers.index,y=top_scorers)
```

__So as we can see the virat kohli is highest run getter in t20 world cup 2022.__

# 5.Analysing How many are the player of the match award in world cup.

```python
player_of_match=data.groupby('player of the match')['player of the match'].count().sort_values(ascending=False)
print(player_of_match)
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation=40)
plt.title('Player of the Match Awards in t20 World Cup 2022')
sns.barplot(x=player_of_match.index,y=player_of_match)
```

__Here we can see Virat Kohli, Taskin Ahmed, Suryakumar Yadav, Shadab Khan and Sam Curran got the player of the match awards in 2 matches. and no more than this players got the 2 player of the match award.__ 

# 6.Analysing the best bowling figure among the bowlers.

```python
best_bowler=data['best bowler'].value_counts()
print(best_bowler)
```

```python
plt.figure(figsize=(20,10))
plt.title('Best Bowlers in t20 World Cup 2022')
plt.xticks(rotation=45)
sns.barplot(x=best_bowler.index,y=best_bowler)
```

__Sam Curran was the only best bowler in 3 macthes.and he was the player of tournament.__

# 7.Analyzing the runs scored in first inning and second inning in every stadium of t20 world cup of 2022.

```python
first_and_second_inning_score=data.groupby('venue').agg({'first innings score':'sum','second innings score':'sum'})
print(first_and_second_inning_score)
```

```python
first_and_second_inning_score.plot(kind='bar',figsize=(20,10),rot=0,title='Best Stadiums to Bat First or Chase',xlabel='Venue',ylabel='Sum of runs')
plt.legend(loc=1)
```

 __So SCG was the best stadium to bat first and followed by Adelaide Oval and in other stadiums there is no much difference while chesing and batting first.__

# 8.Analysing No of wickets lost in first inning and while chesing in every stadium in world cup 2022. 

```python
total_wickets_in_both_innings=data.groupby('venue').agg({'first innings wickets':'sum','second innings wickets':'sum'})
print(total_wickets_in_both_innings)
```

```python
total_wickets_in_both_innings.plot(kind='bar',figsize=(20,5),rot=0,title='Best Statiums to Bowl First or Defend',ylabel='Sum of wickets',xlabel='Venue')
```

__So SCG was the best stadium to bowl while defending the target and Adelaide oval is best to bowl first.__
